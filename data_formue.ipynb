```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Data\\overlevelses_ssh.csv')
over_ssh = np.array(df[(df['aar'] == 2018) & (df['koen'] == 'Mand') & (df['alder'] <100)].survive_r1)
over_ssh[-1] = 0.0

df = pd.read_csv('Data\\formue_cohort.csv')
savings_retirement = df[(df['KOEN']==0) & (df['ALDER']<100)]
liquid_asset = savings_retirement.nettoform_u_pf
illiquid_savings = savings_retirement.FGCX

df = pd.read_csv('Data\\indbetalinger_koen.csv')
contribution_percentage = np.array(df[df['gender'] == "Man"]['indbetalingsprocent'])

age = np.arange(30, 100)
```

```python
import matplotlib.pyplot as plt

# Prettier plotting
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Contribution Percentage (30-64)
axs[0, 0].plot(np.arange(30, 65), contribution_percentage, color='tab:blue', linewidth=2)
axs[0, 0].set_title('Contribution Percentage', fontsize=12, fontweight='bold')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Contribution %')
axs[0, 0].grid(True)

# Liquid Assets
axs[0, 1].plot(age, liquid_asset, color='tab:green', linewidth=2)
axs[0, 1].set_title('Liquid Assets', fontsize=12, fontweight='bold')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Liquid Assets')
axs[0, 1].grid(True)

# Illiquid Assets
axs[1, 0].plot(age, illiquid_savings, color='tab:orange', linewidth=2)
axs[1, 0].set_title('Illiquid Assets', fontsize=12, fontweight='bold')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('Illiquid Assets')
axs[1, 0].grid(True)

# 1 - Survival Probability
axs[1, 1].plot(age, 1 - over_ssh, color='tab:red', linewidth=2)
axs[1, 1].set_title('Death Probability', fontsize=12, fontweight='bold')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Probability')
axs[1, 1].grid(True)

# Tighter layout and better spacing
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
